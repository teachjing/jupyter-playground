```powershell
## You will need to ensure you mount the docker socket connection
## This is for testing purposes. 
## -v /var/run/docker.sock:/var/run/docker.sock

$url = "http://localhost/containers/json"

$response = curl --unix-socket /var/run/docker.sock $url
```

```powershell
$containerList = $response | ConvertFrom-JSON
$containerList.count
```

```powershell
$containerList | Select-Object Image
```

```powershell
## Pull latest image
curl -XPOST --unix-socket /var/run/docker.sock http://localhost/images/create?fromImage=nginx:latest
```

```powershell
## Create an NGINX Container

$url = "http://localhost/containers/create?name=nginx"
$header = @{
    "Content-Type" = "application/json"
}; $header = ($header | ConvertTo-JSON)
$body = @{
       "image" = "nginx"
}; $body = ($body | ConvertTo-JSON); $body

curl -v --unix-socket /var/run/docker.sock --header "Content-Type: application/json" --data $body "http://localhost/containers/create?name=nginx"
```

```powershell
## Get list of containers
$url = "http://localhost/containers/json"
curl --unix-socket /var/run/docker.sock $url
```

```powershell
## Search for image
$response = curl --unix-socket /var/run/docker.sock "http://localhost/images/search?term=nginx"
$response | ConvertFrom-JSON | FT
```
